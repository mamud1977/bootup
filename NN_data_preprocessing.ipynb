```python
# Select specific columns
selected_columns = ["_golden","_unit_state","_last_judgment_at","policies_violated:confidence","city","review"]
df_new = df[selected_columns]
df_new.head(10)
```

```python
## To find the missing values

df_new.isnull().sum()
```

```python
df[df_new["city"].isnull()]
```

```python
df_new["_last_judgment_at"] = pd.to_datetime(df_new["_last_judgment_at"])
## format='%Y-%m-%d %H:%M'

df_new.info()
```

```python
df_new["year"] = df_new["_last_judgment_at"].dt.year
df_new["month"] = df_new["_last_judgment_at"].dt.month

df_new.tail(10)
```

```python
df_new = df_new.drop("_last_judgment_at", axis=1)
df_new.tail(10)
```

```python
df_new[df_new["city"] =="Portland"]
```

```python
df_groupby_city = df_new.groupby("city")
df_groupby_city.get_group("Portland")
```

```python
df_new = df_new.rename(columns={"_golden": "golden","_unit_state":"unit_state","policies_violated:confidence":"confidence"})
df_new.head()
```

```python
unique_values = df_new["city"].unique()

unique_values
```

```python
df_new["city"].value_counts()
```

```python
df_new2 = pd.get_dummies(df_new["city"], dtype=int)
df_new2
```

```python
df_concatenated = pd.concat([df_new, df_new2], axis=1)
df_concatenated
```

```python
df_concatenated.drop("city", axis=1, inplace=True)
df_concatenated
```
