```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)

y = 3*x*x*x - 10*x*x  +4
plt.plot(x,y,'r--')
plt.title('Just a Curve')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

```

```python
mat = np.arange(1,101)
mat = mat.reshape(10,10)

plt.imshow(mat, cmap='coolwarm')
```
